```python
import os
from ultralytics import YOLO
from PIL import Image
import yaml
```

```python
# 1. Paths
BASE_DIR = './../Data'
TRAIN_IMG_DIR = os.path.join(BASE_DIR, 'train_images')
LABEL_DIR = os.path.join(BASE_DIR, 'label')
TEST_IMG_DIR = os.path.join(BASE_DIR, 'test_images')

# dataset structure
dataset_dir = os.path.join(BASE_DIR, 'dataset')
train_img_dst = os.path.join(dataset_dir, 'images/train')
val_img_dst = os.path.join(dataset_dir, 'images/val')
train_lbl_dst = os.path.join(dataset_dir, 'labels/train')
val_lbl_dst = os.path.join(dataset_dir, 'labels/val')
```

```python
# 2. Use existing data.yaml
data_yaml_path = os.path.join(BASE_DIR, 'data.yaml')
if not os.path.isfile(data_yaml_path):
    raise FileNotFoundError(f"data.yaml not found at {data_yaml_path}")
print(f"Using data.yaml at: {data_yaml_path}")
```

```python
# 3. Train YOLOv8 model
model = YOLO('yolov8n.pt')
model.train(data=data_yaml_path, imgsz=640, epochs=50)
```

```python
import os
import json
import pandas as pd
from ultralytics import YOLO

# 1. 클래스 ID 매핑 파일 불러오기,
with open("cls2cat_id_mapping.json", "r", encoding="utf-8") as f:
    classid_to_categoryid = json.load(f)

# 2. 테스트 이미지 파일 정렬,
predict_dir = "./../Data/test_images"
image_files = sorted(
    [f for f in os.listdir(predict_dir) if f.endswith((".png", ".jpg"))],
    key=lambda x: int(os.path.splitext(x)[0])
)

# 3. 결과 저장 리스트,
csv_rows = []
annotation_id = 1

# 4. 이미지별 추론,
for image_file in image_files:
    image_id = int(os.path.splitext(image_file)[0])  # 예: "3.png" → 3
    img_path = os.path.join(predict_dir, image_file)
    results = model(img_path)

    for box in results[0].boxes:
        x1, y1, x2, y2 = box.xyxy[0].tolist()
        bbox_w = x2 - x1
        bbox_h = y2 - y1

        cls_id = int(box.cls[0].item())
        conf = float(box.conf[0].item())
        category_id = int(classid_to_categoryid[str(cls_id)])  # str 키 사용 주의

        csv_rows.append([
            annotation_id,
            image_id,
            category_id,
            round(x1),
            round(y1),
            round(bbox_w),
            round(bbox_h),
            round(conf, 2)
        ])

        annotation_id += 1

# 5. DataFrame으로 변환 및 저장 (컬럼 순서 명시),
columns = [
    "annotation_id", "image_id", "category_id",
    "bbox_x", "bbox_y", "bbox_w", "bbox_h", "score"
]
df = pd.DataFrame(csv_rows, columns=columns)
df.to_csv("submission_formatted.csv", index=False)
print("✅ 제출용 CSV 저장 완료: submission_formatted.csv")
```
